```python
import pandas as pd

item_set = pd.read_csv('pets_smart_store_data_2022_07_01.csv')
item_set
```

```python

columns_to_remove = ['gtin13', 'sku', 'url', 'uniq_id']
item = item_set.drop(columns=columns_to_remove, errors='ignore')

# Save the modified DataFrame back to a CSV file
item.to_csv('itemset.csv', index=False)
item.head()
```

```python
def replace_ampersand_in_list(string_list):
    modified_list = [s.replace("&", " and ") for s in string_list]
    return modified_list

product_names = item['name']
item['name'] = replace_ampersand_in_list(product_names)
item

product_names = item['description']
item['description'] = replace_ampersand_in_list(product_names)
item
```

```python
def convert_usd_to_nrs(amount_usd, exchange_rate=118.75):
    try:
        return float(amount_usd) * exchange_rate
    except ValueError:
        return None  


item['price_in_nrs'] = item['price'].apply(convert_usd_to_nrs)

item.head(100)
```

```python
keywords = ['dog', 'cat', 'food', 'toys']

mask = item['name'].str.contains('|'.join(keywords), case=False)

filtered_item = item[mask]

filtered_item
```

```python
# Function to convert stock status to bool (1 for InStock, 0 for OutOfStock)
def convert_stock_status_to_bool(dataframe, stock_check):
    dataframe = dataframe.copy()
    dataframe.loc[:, 'availability'] = dataframe[stock_check].apply(lambda x: 1 if x.lower() == 'instock' else 0)
    return dataframe

filtered_item = convert_stock_status_to_bool(filtered_item, 'availability')
filtered_item
```

```python
filtered_item['avg_rating'].fillna(filtered_item['avg_rating'].mean(), inplace=False)
filtered_item
```

```python
# Calculate mean excluding NaN values
mean_rating = filtered_item['avg_rating'].mean(skipna=True)

# Fill NaN values with the calculated mean
filtered_item['avg_rating'] = filtered_item['avg_rating'].fillna(mean_rating)
filtered_item
```

```python
# Replace 'price_in_nrs' with 'price' and rename the column
filtered_item['price'] = filtered_item['price_in_nrs']
filtered_item.drop(['currency', 'reviews_count', 'scraped_at', 'price_in_nrs'], axis=1, inplace=True)
filtered_item.rename(columns={'price': 'price (NRs)'}, inplace=True)
```

```python
filtered_item.to_csv('itemset.csv', index=False)
filtered_item
```
